Seus Dados agora

```python
from trilha_python.user_profile import UserProfile

def coletar_dados_interativo():
    """
    Função de Interface: Responsável apenas pela captura dos dados brutos.
    Separa a entrada (Input) do processamento (Business Logic).
    """
    print(f"{'='*40}")
    print("      CADASTRO DE PERFIL - 2026")
    print(f"{'='*40}")

    return {
        "nome": input("👉 Digite o nome completo: "),
        "idade": input("👉 Digite a idade: "),
        "endereco": input("👉 Digite o endereço: "),
        "altura": input("👉 Digite a altura (ex: 1.78): "),
        "peso": input("👉 Digite o peso em kg (ex: 80.5): "),
        "sexo": input("👉 Digite o sexo (M/F/O): "),
    }

# --- Fluxo de Execução Principal ---
if __name__ == "__main__":
    sucesso = False

    while not sucesso:
        try:
            # 1. Captura os dados via teclado
            dados_digitados = coletar_dados_interativo()

            # 2. Tenta instanciar a classe (onde ocorre a validação)
            usuario = UserProfile(dados_digitados)

            # 3. Se chegou aqui, os dados são válidos
            print(usuario.gerar_relatorio())
            sucesso = True

        except ValueError as e:
            # Captura erros de validação (ex: letra na idade ou campo vazio)
            print(f"\n⚠️ {e}")
            print("Por favor, tente preencher os dados novamente.\n")

        except Exception as e:
            print(f"\n💥 Ocorreu um erro inesperado: {e}")
            break
```

Exercício

```python
# --- 1. Criando a Lista com 3 números digitados pelo usuário ---
lista_numeros = []
print("Fase 1: Preenchendo a Lista")
for i in range(3):
    num = int(input(f"Digite o {i+1}º número inteiro: "))
    lista_numeros.append(num)

# --- 2. Criando a Tupla com 5 números inteiros ---
# Tuplas são imutáveis, ideais para valores fixos de configuração.
tupla_fixa = (10, 20, 30, 40, 50)

# --- 3. Criando o Dicionário de 3 pessoas com suas idades ---
# Aqui, usamos nomes como chaves e idades como valores.
dicionario_pessoas = {
    "Vitor": 37,    # Sua idade atual conforme os registros
    "Alice": 25,
    "Marcos": 42
}

# --- 4. Mostrando todos os dados ---
print("\n" + "="*30)
print("📊 RESULTADOS DO EXERCÍCIO")
print("="*30)
print(f"✅ Lista (Dinâmica):    {lista_numeros}")
print(f"✅ Tupla (Estática):    {tupla_fixa}")
print(f"✅ Dicionário (Pares):  {dicionario_pessoas}")
print("="*30)
```

Fatorial

```python
numero = int (input('Digite um número: '))
f = 1
for i in range(numero, 1, -1):
    f = f * i
print(f'O fatorial de {numero} é {f}')
```
