# Seus Dados agora

```python
from trilha_python.user_profile import UserProfile

def coletar_dados_interativo():
    """
    Função de Interface: Responsável apenas pela captura dos dados brutos.
    Separa a entrada (Input) do processamento (Business Logic).
    """
    print(f"{'='*40}")
    print("      CADASTRO DE PERFIL - 2026")
    print(f"{'='*40}")

    return {
        "nome": input("👉 Digite o nome completo: "),
        "idade": input("👉 Digite a idade: "),
        "endereco": input("👉 Digite o endereço: "),
        "altura": input("👉 Digite a altura (ex: 1.78): "),
        "peso": input("👉 Digite o peso em kg (ex: 80.5): "),
        "sexo": input("👉 Digite o sexo (M/F/O): "),
    }

# --- Fluxo de Execução Principal ---
if __name__ == "__main__":
    sucesso = False

    while not sucesso:
        try:
            # 1. Captura os dados via teclado
            dados_digitados = coletar_dados_interativo()

            # 2. Tenta instanciar a classe (onde ocorre a validação)
            usuario = UserProfile(dados_digitados)

            # 3. Se chegou aqui, os dados são válidos
            print(usuario.gerar_relatorio())
            sucesso = True

        except ValueError as e:
            # Captura erros de validação (ex: letra na idade ou campo vazio)
            print(f"\n⚠️ {e}")
            print("Por favor, tente preencher os dados novamente.\n")

        except Exception as e:
            print(f"\n💥 Ocorreu um erro inesperado: {e}")
            break
```

# While True try

```python
numeros = []

while True:
    try:
        entrada = input('Digite um número (ou negativo para sair): ')
        n = int(entrada)

        if n < 0:
            break

        numeros.append(n)
    except ValueError:
        print("Ops! Por favor, digite apenas números inteiros.")

# A "mágica" do Python: soma tudo de uma vez
soma = sum(numeros)

print(f"\nLista final: {numeros}")
print(f"Soma total: {soma}")
```

# **Desafio Descomplica: Verificador de Número Primo **

```python
while True:
    try:
        entrada = input("\nDigite um número inteiro positivo (ou 'sair'): ")

        if entrada.lower() == 'sair':
            print("Encerrando o sistema... Até mais! 🔨")
            break

        num = int(entrada)

        if num < 0:
            print("⚠️ Por favor, digite um número positivo.")
            continue

        if num <= 1:
            print(f"❌ O número {num} não é primo.")
            continue

        e_primo = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                e_primo = False
                break

        if e_primo:
            print(f"✅ O número {num} é PRIMO!")
        else:
            print(f"❌ O número {num} NÃO é primo.")

    except ValueError:

        print("⚠️ Erro: Entrada inválida! Digite apenas números inteiros.")
```

```python
def obter_maior(matriz):
    """Retorna o maior valor dentro da matriz 3x3"""
    maior = matriz[0][0]
    for linha in matriz:
        for num in linha:
            if num > maior:
                maior = num
    return maior

def obter_soma(matriz):
    """Retorna a soma de todos os elementos da matriz"""
    soma = 0
    for linha in matriz:
        soma += sum(linha)
    return soma

# --- PROGRAMA PRINCIPAL ---

matriz = []

print("--- Entrada de Dados (9 números) ---")
for i in range(3):
    linha = []
    for j in range(3):
        valor = int(input(f"Digite o valor para a posição [{i}][{j}]: "))
        linha.append(valor)
    matriz.append(linha)

# Executando as funções modulares
resultado_maior = obter_maior(matriz)
resultado_soma = obter_soma(matriz)

print("\n" + "="*20)
print(f"📊 MATRIZ GERADA: {matriz}")
print(f"🔝 MAIOR NÚMERO: {resultado_maior}")
print(f"➕ SOMA TOTAL: {resultado_soma}")
print("="*20)
```

Verifica se um número não é primo

```python
# --- MÓDULO DE LÓGICA ---

def verificar_se_e_primo(n):
    """Verifica se um número n é primo."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# --- PROGRAMA PRINCIPAL ---

vetor_numeros = []

print("--- Entrada de Dados (10 números) ---")
for i in range(10):
    num = int(input(f"Digite o {i+1}º número inteiro: "))
    vetor_numeros.append(num)

print("\n--- Resultado da Verificação ---")
for numero in vetor_numeros:
    if verificar_se_e_primo(numero):
        print(f"O número {numero:2} -> É PRIMO ✅")
    else:
        print(f"O número {numero:2} -> NÃO é primo ❌")
```
